```python
#Detecting and showing the face in the image, Cleanup the useless images
def cleanup_images(actress_name):
    folder_location = "../Celebs/{}/*.jpg".format(actress_name)
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
    actress_images = np.sort(np.array(glob(folder_location)))
    image_counter = 0
    for image_str in actress_images:
        try:
            img = cv2.imread(image_str)
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            faces = face_cascade.detectMultiScale(gray)
            if len(faces) == 1:
                image_counter += 1
            else:
                os.remove(image_str)
        except:
            os.remove(image_str)

    print("Out of {} images we could capture {} single faces".format(len(actress_images), image_counter))   
```

```python
cleanup_images("deepika")
```
